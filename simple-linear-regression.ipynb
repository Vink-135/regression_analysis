```python
dataset = pd.read_csv(r"/kaggle/input/cgpa-vs-package-in-lpa/placement.csv")
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
x = dataset[["cgpa"]]
y = dataset["package"]
x.ndim
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```
